### Quick Sort Algorithm

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import time

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)

# 사용자에게 데이터 개수 입력 받기
data_size = int(input("정렬할 데이터의 개수를 입력하세요: "))

# 1부터 1001 사이의 랜덤한 정수를 data_size 개 생성
data = np.random.randint(1, 1001, size=data_size)
initial_data = data.copy()

# 퀵 정렬 실행
start = time.time()
sorted_data = quick_sort(data)
end = time.time()

# 초기 상태와 최종 상태를 배열로 출력
print("초기 데이터:", initial_data)
print("정렬된 데이터:", sorted_data)
print(f"{end - start:.5f} sec")

# 초기 상태와 최종 상태 비교 (그래프)
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.bar(range(len(initial_data)), initial_data, color='blue')
plt.title("Initial State")
plt.subplot(1, 2, 2)
plt.bar(range(len(sorted_data)), sorted_data, color='green')
plt.title("Final State")
plt.show()
```

### Bubble Sort Algorithm

```python
import matplotlib.pyplot as plt
import numpy as np

# 버블 정렬 함수
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

# 사용자에게 데이터 개수 입력 받기
data_size = int(input("정렬할 데이터의 개수를 입력하세요: "))

# 1부터 1000 사이의 랜덤한 정수를 data_size 개 생성
data = np.random.randint(1, 1001, size=data_size)
initial_data = data.copy()

# 버블 정렬 실행
start = time.time()
sorted_data = bubble_sort(data)
end = time.time()

# 초기 상태와 최종 상태를 배열로 출력
print("초기 데이터:", initial_data)
print("정렬된 데이터:", sorted_data)
print(f"{end - start:.5f} sec")

# 초기 상태와 최종 상태 비교 (그래프)
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.bar(range(len(initial_data)), initial_data, color='blue')
plt.title("Initial State")
plt.subplot(1, 2, 2)
plt.bar(range(len(sorted_data)), sorted_data, color='green')
plt.title("Final State")
plt.show()
```

### Merge Sort Algorithm

```python
import matplotlib.pyplot as plt
import numpy as np
import time

# 합병 정렬 함수
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    
    return merge(left, right)

# 병합 함수
def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

# 사용자에게 데이터 개수 입력 받기
data_size = int(input("정렬할 데이터의 개수를 입력하세요: "))

# 1부터 1000 사이의 랜덤한 정수를 data_size 개 생성
data = np.random.randint(1, 1001, size=data_size)
initial_data = data.copy()

# 합병 정렬 실행
start = time.time()
sorted_data = merge_sort(data)
end = time.time()

# 초기 상태와 최종 상태를 배열로 출력
print("초기 데이터:", initial_data)
print("정렬된 데이터:", sorted_data)
print(f"{end - start:.5f} sec")

# 초기 상태와 최종 상태 비교 (그래프)
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.bar(range(len(initial_data)), initial_data, color='blue')
plt.title("Initial State")
plt.subplot(1, 2, 2)
plt.bar(range(len(sorted_data)), sorted_data, color='green')
plt.title("Final State")
plt.show()
```

# 새 창 띄워서 정렬 시각화

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import tkinter as tk
from tkinter import simpledialog
import time  # 실행 시간 측정을 위한 모듈

# 퀵 정렬 함수
def quick_sort(arr, start, end, draw_data, canvas, ax, delay, back_button, window):
    if start < end:
        pivot_index = partition(arr, start, end)
        draw_data(arr, canvas, ax)
        root.update()  # tkinter 창 업데이트
        root.after(int(delay * 1000))  # 지연시간

        quick_sort(arr, start, pivot_index - 1, draw_data, canvas, ax, delay, back_button, window)
        quick_sort(arr, pivot_index + 1, end, draw_data, canvas, ax, delay, back_button, window)

    if back_button.winfo_exists():
        back_button.config(state=tk.NORMAL)  # 정렬이 끝나면 버튼 활성화

    # 마지막에 finish_sorting 호출
    if start == 0 and end == len(arr) - 1:  # 전체 정렬이 끝났을 때만 호출
        finish_sorting(window, arr)  # 정렬 완료 후 처리

def partition(arr, start, end):
    pivot = arr[end]
    i = start - 1
    for j in range(start, end):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[end] = arr[end], arr[i + 1]
    return i + 1

# 버블 정렬 함수
def bubble_sort(arr, draw_data, canvas, ax, delay, back_button, window):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
        draw_data(arr, canvas, ax)
        root.update()  # tkinter 창 업데이트
        root.after(int(delay * 1000))  # 지연시간

    if back_button.winfo_exists():
        back_button.config(state=tk.NORMAL)  # 정렬이 끝나면 버튼 활성화
    finish_sorting(window, arr)  # 정렬 완료 후 처리

# 합병 정렬 함수
def merge_sort(arr, start, end, draw_data, canvas, ax, delay, back_button, window):
    if start < end - 1:  # 두 개 이상의 요소가 있을 때만 계속 분할
        mid = (start + end) // 2
        merge_sort(arr, start, mid, draw_data, canvas, ax, delay, back_button, window)
        merge_sort(arr, mid, end, draw_data, canvas, ax, delay, back_button, window)
        merge(arr, start, mid, end, draw_data, canvas, ax, delay)

    # 마지막 병합 후 완료 처리
    if start == 0 and end == len(arr):
        if back_button.winfo_exists():
            back_button.config(state=tk.NORMAL)  # 정렬이 끝나면 버튼 활성화
        finish_sorting(window, arr)  # 정렬 완료 후 처리

# 병합 함수
def merge(arr, start, mid, end, draw_data, canvas, ax, delay):
    left = arr[start:mid].copy()  # 왼쪽 부분 배열 복사
    right = arr[mid:end].copy()  # 오른쪽 부분 배열 복사
    i = j = 0
    k = start

    # 좌우 배열을 병합하면서 원본 배열에 다시 저장
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:  # 등호 포함으로 안정적 정렬 유지
            arr[k] = left[i]
            i += 1
        else:
            arr[k] = right[j]
            j += 1
        k += 1

    # 왼쪽 배열에 남은 요소들을 복사
    while i < len(left):
        arr[k] = left[i]
        i += 1
        k += 1

    # 오른쪽 배열에 남은 요소들을 복사
    while j < len(right):
        arr[k] = right[j]
        j += 1
        k += 1

    # 병합된 결과를 화면에 그리기
    draw_data(arr, canvas, ax)
    root.update()  # tkinter 창 업데이트
    root.after(int(delay * 1000))  # 지연시간 추가



# 그래프 시각화 함수
def draw_data(data, canvas, ax, color='blue'):
    ax.clear()
    ax.bar(range(len(data)), data, color=color)
    ax.set_title("Sorting Visualization")
    canvas.draw()

# 정렬이 완료되었을 때 창을 닫고 프로그램이 종료되지 않도록 처리
def finish_sorting(window, sorted_data):
    end_time = time.time()  # 정렬 완료 시 시간 기록
    sorting_time = end_time - start_time  # 총 정렬 시간 계산

    # 정렬 전후 배열 및 그래프 출력
    print("정렬 전 데이터:", original_data)
    print("정렬 후 데이터:", sorted_data)
    print(f"총 정렬 시간: {sorting_time:.2f} 초")
    show_comparison_graph(original_data, sorted_data)

# 정렬 전후 배열을 시각적으로 비교하기 위한 그래프 출력
def show_comparison_graph(original_data, sorted_data):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    ax1.bar(range(len(original_data)), original_data, color='blue')
    ax1.set_title("Original Data")

    ax2.bar(range(len(sorted_data)), sorted_data, color='green')
    ax2.set_title("Sorted Data")

    plt.show()

# GUI에서 데이터를 입력받아 실행하는 함수
def start_sorting():
    global original_data, start_time, root

    # Tkinter 창을 통해 데이터 크기 입력 받기
    data_size = simpledialog.askinteger("Input", "정렬할 데이터의 개수를 입력하세요:", minvalue=5, maxvalue=1000)

    # 랜덤 데이터 생성 및 정렬 전 데이터 저장
    original_data = np.random.randint(1, 10001, size=data_size)

    # 정렬 선택
    sort_choice = simpledialog.askstring("Input", "퀵정렬(Q), 버블정렬(B), 합병정렬(M) 중 하나를 선택하세요:")

    # 새로운 창에서 그래프 시각화
    window = tk.Toplevel(root)
    window.title("Sorting Visualization")

    fig, ax = plt.subplots(figsize=(10, 6))
    canvas = FigureCanvasTkAgg(fig, master=window)
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=True)

    # 다시 처음 화면으로 돌아가는 버튼 추가
    back_button = tk.Button(window, text="처음으로 돌아가기", state=tk.DISABLED, command=lambda: reset_window(window))
    back_button.pack(side=tk.BOTTOM)

    # 정렬 시작 시간 기록
    global start_time
    start_time = time.time()

    # 정렬 방식에 따라 분기 처리
    if sort_choice.lower() == 'q':
        quick_sort(original_data.copy(), 0, len(original_data) - 1, lambda x, c, a: draw_data(x, c, a), canvas, ax, delay=0.1, back_button=back_button, window=window)
    elif sort_choice.lower() == 'b':
        bubble_sort(original_data.copy(), lambda x, c, a: draw_data(x, c, a), canvas, ax, delay=0.1, back_button=back_button, window=window)
    elif sort_choice.lower() == 'm':
        merge_sort(original_data.copy(), 0, len(original_data), lambda x, c, a: draw_data(x, c, a), canvas, ax, delay=0.1, back_button=back_button, window=window)
    else:
        print("올바른 정렬 알고리즘을 선택하세요 (Q, B, M).")

# 창 닫기 및 메인화면 복귀 함수
def reset_window(window):
    window.destroy()  # 현재 창 닫기
    start_sorting()  # 다시 시작

# Tkinter로 실행창 띄우기
if __name__ == "__main__":
    root = tk.Tk()
    root.withdraw()  # 메인 창 숨기기
    start_sorting()
    root.mainloop()  # 메인 루프 실행
```

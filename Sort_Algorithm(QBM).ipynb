### Quick Sort Algorithm

```python
import matplotlib.pyplot as plt
import numpy as np

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)

# 사용자에게 데이터 개수 입력 받기
data_size = int(input("정렬할 데이터의 개수를 입력하세요: "))

# 1부터 1000 사이의 랜덤한 정수를 data_size 개 생성
data = np.random.randint(1, 1001, size=data_size)
initial_data = data.copy()

# 퀵 정렬 실행
sorted_data = quick_sort(data)

# 초기 상태와 최종 상태를 배열로 출력
print("초기 데이터:", initial_data)
print("정렬된 데이터:", sorted_data)

# 초기 상태와 최종 상태 비교 (그래프)
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.bar(range(len(initial_data)), initial_data, color='blue')
plt.title("Initial State")
plt.subplot(1, 2, 2)
plt.bar(range(len(sorted_data)), sorted_data, color='green')
plt.title("Final State")
plt.show()
```

### Bubble Sort Algorithm

```python
import matplotlib.pyplot as plt
import numpy as np

# 버블 정렬 함수
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

# 사용자에게 데이터 개수 입력 받기
data_size = int(input("정렬할 데이터의 개수를 입력하세요: "))

# 1부터 1000 사이의 랜덤한 정수를 data_size 개 생성
data = np.random.randint(1, 1001, size=data_size)
initial_data = data.copy()

# 버블 정렬 실행
sorted_data = bubble_sort(data)

# 초기 상태와 최종 상태를 배열로 출력
print("초기 데이터:", initial_data)
print("정렬된 데이터:", sorted_data)

# 초기 상태와 최종 상태 비교 (그래프)
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.bar(range(len(initial_data)), initial_data, color='blue')
plt.title("Initial State")
plt.subplot(1, 2, 2)
plt.bar(range(len(sorted_data)), sorted_data, color='green')
plt.title("Final State")
plt.show()
```

### Merge Sort Algorithm

```python
import matplotlib.pyplot as plt
import numpy as np

# 합병 정렬 함수
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    
    return merge(left, right)

# 병합 함수
def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

# 사용자에게 데이터 개수 입력 받기
data_size = int(input("정렬할 데이터의 개수를 입력하세요: "))

# 1부터 1000 사이의 랜덤한 정수를 data_size 개 생성
data = np.random.randint(1, 1001, size=data_size)
initial_data = data.copy()

# 합병 정렬 실행
sorted_data = merge_sort(data)

# 초기 상태와 최종 상태를 배열로 출력
print("초기 데이터:", initial_data)
print("정렬된 데이터:", sorted_data)

# 초기 상태와 최종 상태 비교 (그래프)
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.bar(range(len(initial_data)), initial_data, color='blue')
plt.title("Initial State")
plt.subplot(1, 2, 2)
plt.bar(range(len(sorted_data)), sorted_data, color='green')
plt.title("Final State")
plt.show()
```

### 새 창 띄워서 정렬 시각화

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import tkinter as tk
from tkinter import simpledialog

# 퀵 정렬 함수
def quick_sort(arr, start, end, draw_data, canvas, ax, delay, back_button):
    if start < end:
        pivot_index = partition(arr, start, end)
        draw_data(arr, canvas, ax)
        root.update()  # tkinter 창 업데이트
        root.after(int(delay * 1000))  # 지연시간

        quick_sort(arr, start, pivot_index - 1, draw_data, canvas, ax, delay, back_button)
        quick_sort(arr, pivot_index + 1, end, draw_data, canvas, ax, delay, back_button)
    else:
        back_button.config(state=tk.NORMAL)  # 정렬이 끝나면 버튼 활성화

def partition(arr, start, end):
    pivot = arr[end]
    i = start - 1
    for j in range(start, end):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[end] = arr[end], arr[i+1]
    return i + 1

# 버블 정렬 함수
def bubble_sort(arr, draw_data, canvas, ax, delay, back_button):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
        draw_data(arr, canvas, ax)
        root.update()  # tkinter 창 업데이트
        root.after(int(delay * 1000))  # 지연시간

    back_button.config(state=tk.NORMAL)  # 정렬이 끝나면 버튼 활성화

# 합병 정렬 함수
def merge_sort(arr, start, end, draw_data, canvas, ax, delay, back_button):
    if end - start > 1:
        mid = (start + end) // 2
        merge_sort(arr, start, mid, draw_data, canvas, ax, delay, back_button)
        merge_sort(arr, mid, end, draw_data, canvas, ax, delay, back_button)
        merge(arr, start, mid, end, draw_data, canvas, ax, delay)

    back_button.config(state=tk.NORMAL)  # 정렬이 끝나면 버튼 활성화

def merge(arr, start, mid, end, draw_data, canvas, ax, delay):
    left = arr[start:mid]
    right = arr[mid:end]
    i = j = 0
    k = start

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            arr[k] = left[i]
            i += 1
        else:
            arr[k] = right[j]
            j += 1
        k += 1

    while i < len(left):
        arr[k] = left[i]
        i += 1
        k += 1

    while j < len(right):
        arr[k] = right[j]
        j += 1
        k += 1

    draw_data(arr, canvas, ax)
    root.update()  # tkinter 창 업데이트
    root.after(int(delay * 1000))  # 지연시간

# 그래프 시각화 함수
def draw_data(data, canvas, ax, color='blue'):
    ax.clear()
    ax.bar(range(len(data)), data, color=color)
    ax.set_title("Sorting Visualization")
    canvas.draw()

# GUI에서 데이터를 입력받아 실행하는 함수
def start_sorting():
    global root

    # Tkinter 창을 통해 데이터 크기 입력 받기
    data_size = simpledialog.askinteger("Input", "정렬할 데이터의 개수를 입력하세요:", minvalue=5, maxvalue=1000)

    # 랜덤 데이터 생성
    data = np.random.randint(1, 10001, size=data_size)

    # 정렬 선택
    sort_choice = simpledialog.askstring("Input", "퀵정렬(Q), 버블정렬(B), 합병정렬(M) 중 하나를 선택하세요:")

    # 새로운 창에서 그래프 시각화
    window = tk.Toplevel(root)
    window.title("Sorting Visualization")

    fig, ax = plt.subplots(figsize=(10, 6))
    canvas = FigureCanvasTkAgg(fig, master=window)
    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=True)

    # 다시 처음 화면으로 돌아가는 버튼 추가
    back_button = tk.Button(window, text="처음으로 돌아가기", state=tk.DISABLED, command=lambda: reset_window(window))
    back_button.pack(side=tk.BOTTOM)

    # 정렬 방식에 따라 분기 처리
    if sort_choice.lower() == 'q':
        quick_sort(data, 0, len(data) - 1, lambda x, c, a: draw_data(x, c, a), canvas, ax, delay=0.1, back_button=back_button)
    elif sort_choice.lower() == 'b':
        bubble_sort(data, lambda x, c, a: draw_data(x, c, a), canvas, ax, delay=0.1, back_button=back_button)
    elif sort_choice.lower() == 'm':
        merge_sort(data, 0, len(data), lambda x, c, a: draw_data(x, c, a), canvas, ax, delay=0.1, back_button=back_button)
    else:
        print("올바른 정렬 알고리즘을 선택하세요 (Q, B, M).")

# 창 닫기 및 메인화면 복귀 함수
def reset_window(window):
    window.destroy()  # 현재 창 닫기
    start_sorting()  # 다시 시작

# Tkinter로 실행창 띄우기
if __name__ == "__main__":
    root = tk.Tk()
    root.withdraw()  # 메인 창 숨기기
    start_sorting()
```

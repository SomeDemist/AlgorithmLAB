```python
import matplotlib.pyplot as plt
import numpy as np

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)

# 사용자에게 데이터 개수 입력 받기
data_size = int(input("정렬할 데이터의 개수를 입력하세요: "))

# 1부터 1000 사이의 랜덤한 정수를 data_size 개 생성
data = np.random.randint(1, 1001, size=data_size)
initial_data = data.copy()

# 퀵 정렬 실행
sorted_data = quick_sort(data)

# 초기 상태와 최종 상태를 배열로 출력
print("초기 데이터:", initial_data)
print("정렬된 데이터:", sorted_data)

# 초기 상태와 최종 상태 비교 (그래프)
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.bar(range(len(initial_data)), initial_data, color='blue')
plt.title("Initial State")
plt.subplot(1, 2, 2)
plt.bar(range(len(sorted_data)), sorted_data, color='green')
plt.title("Final State")
plt.show()
```
